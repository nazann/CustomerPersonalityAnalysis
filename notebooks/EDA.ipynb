### We culd see by increase of Income spending goes almost like exponentially. Iam quite curious about where is the most dramatic jump.

```python
df_sorted = df_filtered.sort_values(by='Income').reset_index(drop=True)
df_sorted.head()
```

```python
df_sorted['Spending_Pct_Increase'] = df_sorted['Total_Spending'].pct_change()
df_sorted['Income_Diff'] = df_sorted['Income'].diff()
```

```python
max_idx = df_sorted['Spending_Pct_Increase'].idxmax()
max_jump = df_sorted.loc[max_idx]
prev_row = df_sorted.loc[max_idx - 1]
```

```python
print(f"Highest spending % increase: {max_jump['Spending_Pct_Increase']:.2%}")
print(f"Income jump: from {prev_row['Income']} to {max_jump['Income']}")
print(f"Spending jump: from {prev_row['Total_Spending']} to {max_jump['Total_Spending']}")
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df_sorted['Income'], df_sorted['Spending_Pct_Increase'], marker='o', alpha=0.6)
plt.title('Percentage Increase in Spending Between Income Levels')
plt.xlabel('Income')
plt.ylabel('Spending % Increase')
plt.axvline(max_jump['Income'], color='red', linestyle='--', label='Max Jump')
plt.legend()
plt.show()
```

# Key Findings in correlation Tb:

### Catalog and Store Channels Dominate Spending Behavior:
 Spending on all product categories shows a strong positive correlation with the number of catalog and store purchases.
 Example: Meat Products spending correlates strongly with Catalog Purchases *(correlation coefficient ≈ 0.73)* and Store Purchases (≈ 0.49).
  Wine spending also shows strong links with Catalog (≈ 0.63) and Store Purchases (≈ 0.64).

### Web Purchases Show Moderate Positive Correlation:
    Spending on most products moderately correlates with web purchases, especially for wines (≈ 0.55) and gold products (≈ 0.41).
    This suggests customers who spend more on these products tend to purchase them online.

### Deals Purchases Have Little Impact:
    Number of deal purchases has very weak or negligible correlation with product spending, indicating that buying through deals may not significantly affect total spending.

### Negative Correlation with Web Visits:
    A surprising finding is the negative correlation between web visits per month and product spending (coefficients range from about -0.25 to -0.54).
    This may imply that frequent site visitors are not necessarily heavy spenders, possibly browsing without making many purchases.

# Conclusion:

The analysis highlights that catalog and physical store purchases are the strongest drivers of spending across product categories in this dataset. Web purchases also contribute notably for some products like wines and gold. Deal-based buying and frequent website visits do not strongly influence spending levels.

These insights can help marketing and sales teams optimize channel focus, for example by boosting catalog and store promotions for meat and wine products, or enhancing online sales strategies for gold items.
